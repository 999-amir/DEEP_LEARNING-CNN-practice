# import

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
import tensorflow as tf
```

# load dataset

```python
(X_train, Y_train), (X_test, Y_test) = cifar10.load_data()
```

```python
X = X_train[:10]
Y = Y_train[:10]
```

```python
X.shape
```

```python
plt.imshow(X[0]).figure.set_size_inches(3, 3)
```

# normalize

```python
X = X/X.max()
```

```python
X = X.astype('float')
```

# create filter

```python
filters = np.zeros((3, 3, 3, 1))
```

```python
for i in range(3):
    filters[i, i, :, :] = 1
    filters[i, 2-i, :, :] = 1
```

```python
plt.imshow(filters[:, :, :, 0]).figure.set_size_inches(3, 3)
```

# filter image

```python
convolved_img = tf.nn.conv2d(X, filters, strides=1, padding='SAME')
```

```python
plt.imshow(convolved_img[1, :, :, :]).figure.set_size_inches(3, 3)
```

# effect of filter on image

```python
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(17, 5))
j, k = 0, 0
for i in range(10):
    if k >= 6:
        k -= 6
        j += 1
    ax[j, k].imshow(X[i])
    k += 1
```

```python
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(17, 5))
j, k = 0, 0
for i in range(10):
    if k >= 6:
        k -= 6
        j += 1
    ax[j, k].imshow(convolved_img[i])
    k += 1
```
